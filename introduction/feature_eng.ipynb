# Missing Value Types

## Missing Completely at Random (MCAR)
- Mİssing value occurs randomly without a relationship with any variable in the dataset
- Ex: Someone forgets to fill in their age

## Missing at Random (MAR)

- Missing Values might be about the other variables
- Ex: Man are likely to skip answering questions about weight

## Missing Not at Random (MNAR)

- Missing values depend on the unobserved data itself reason for missingness is not Randon.
- Rich people don't want to report thei income.

```python
np.random.seed(42)

set1no = 900
set2no = 100
```

```python
df1 = pd.DataFrame(
    {
        "feature_1": np.random.normal(loc=0, scale=1, size=set1no),
        "feature_2": np.random.normal(loc=0, scale=1, size=set1no),
        "target": [0] * set1no
    }
)


df2 = pd.DataFrame(
    {
        "feature_1": np.random.normal(loc=0, scale=1, size=set2no),
        "feature_2": np.random.normal(loc=0, scale=1, size=set2no),
        "target": [1] * set2no
    }
)
```

```python
df3 = pd.concat([df1, df2]).reset_index(drop=True)
```

```python
df3
```

```python
df3["target"].unique()
```

```python
df3["target"].value_counts()
```

# Upsampling & Downsampling 

```python
# upsampling -> upsample minority
df_minority = df3[df3["target"] == 1]
# downsampling -> downsample majority
df_majority = df3[df3["target"] == 0]
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsampled = resample(df_minority, replace=True, n_samples=len(df_majority), random_state=42)

df_upsampled = pd.concat([df_minority_upsampled, df_majority])

df_upsampled["target"].value_counts()
```

```python
df_majority_downsampled = resample(df_majority, replace=True, n_samples=len(df_minority), random_state=42)

df_downsampled = pd.concat([df_majority_downsampled, df_minority])

df_downsampled["target"].value_counts()
```

# Data Encoding

```python
# one hot encoding

colors_df = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue', 'Green', 'Red', 'Blue', 'Green', 'Red', 'Blue', 'Green']})
df_encoded = pd.get_dummies(colors_df, columns=['Color'])
df_encoded
```

```python
# Label Encoding -> convert categorical data into numerical data
# its useful when categorical data has ranking or ordered 
# can be used for education level, income level, etc.

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue', 'Green', 'Red', 'Blue', 'Green', 'Red', 'Blue', 'Green']})
df['Color_encoded'] = encoder.fit_transform(df['Color'])
df
```

```python
# Ordinal Encoding -> convert categorical data into numerical data
# its useful when categorical data has ranking or ordered 
# can be used for education level, income level, etc.

from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()

df = pd.DataFrame({'Education': ['High School', 'Bachelors', 'Masters', 'High School', 'Masters', 'Bachelors', 'High School', 'Masters', 'Bachelors', 'High School', 'Masters', 'Bachelors']})
df['Education_encoded'] = encoder.fit_transform(df[['Education']])
df
```
